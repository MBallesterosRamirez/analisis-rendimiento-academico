```python
#Importar df. mostrar encabezados
import pandas as pd
df = pd.read_excel("datos_limpios_resu_P1_1003.xlsx")
df.head()
```

```python
#PASO 2 — QUITAMOS TILDES.. de encabezados.

import unicodedata

def quitar_tildes(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )

df.columns = [quitar_tildes(col) for col in df.columns]
#df.columns
```

```python
#PASO 3 _Revisamos (columnas vacias)
[col for col in df.columns if "Unnamed" in col]
```

```python
#Eliminamos (columnas vacias)
df = df.drop(columns=['Unnamed: 0', 'Unnamed: 37'])
```

```python
df.columns
```

```python
#PASO 4 — NORMALIZAR NOMBRES
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("__", "_")
)
```

```python
df.columns
```

```python
#4.1  (Eliminar doble _ _ y reemplazar por _ )
df.columns = df.columns.str.replace("__", "_")
df.columns
```

```python
import pandas as pd

df_notas = pd.read_csv("datos_notas.csv")
df_convivencia = pd.read_csv("datos_convivencia.csv")
```

```python
df_convivencia.tail()
```

```python
df_notas.tail(2)
```

```python
#PASO 7 — TRANSFORMAR A FORMATO LARGO
df_largo = df_notas.melt(
    id_vars=['codigo', 'estudiante'],
    var_name='asignatura_tipo',
    value_name='nota'
)
```

```python
df_largo.head(2)
```

```python
#PASO 8 — SEPARAR COLUMNA ejempo : jv1, acum
df_largo[['asignatura', 'tipo']] = df_largo['asignatura_tipo'].str.rsplit('_', n=1, expand=True)
```

```python
df_largo.tail(1)
```

```python
#PASO 9 — Eliminación de columnas innecesarias : asignatura_tipo
df_largo = df_largo.drop(columns=['asignatura_tipo'])
```

```python
df_largo.tail(1)
```

```python
#PASO 10 -  DATSET FINAL........................
df_modelo = df_largo.pivot_table(
    index=['codigo', 'estudiante', 'asignatura'],
    columns='tipo',
    values='nota'
).reset_index()
```

```python
#Organizar columnas
df_modelo = df_modelo[['codigo', 'estudiante', 'asignatura', 'jv1', 'acum']]
```

```python
df_modelo.columns.name = None
```

```python
df_modelo.tail(2)
```

```python
df_convivencia.tail(2)
```

```python
df_largo.to_csv("datos_notas_flargo.csv", index=False)
df_convivencia.to_csv("datos_convivencia.csv", index=False)
df_modelo.to_csv("datos_notas_flimpio.csv", index=False)
```
